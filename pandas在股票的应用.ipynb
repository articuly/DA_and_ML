```python
import pandas as pd
import numpy as np
# 读取csv文件，股票代号存在以000开头的情况，为了避免pandas自动转为整数，这里手动指定数据为string格式
file1 = r'data_set/3-9 concept_list.csv'
concept = pd.read_csv(file1, dtype=np.str, encoding='utf-8')
concept.head()
```

```python
file2 = r'data_set/3-9 stock_20190902.csv'
file3 = r'data_set/3-9 stock_20190903.csv'
file4 = r'data_set/3-9 stock_20190904.csv'
file5 = r'data_set/3-9 stock_20190905.csv'
file6 = r'data_set/3-9 stock_20190906.csv'
st902 = pd.read_csv(file2, dtype=np.str, encoding='utf-8')
st903 = pd.read_csv(file3, dtype=np.str, encoding='utf-8')
st904 = pd.read_csv(file4, dtype=np.str, encoding='utf-8')
st905 = pd.read_csv(file5, dtype=np.str, encoding='utf-8')
st906 = pd.read_csv(file6, dtype=np.str, encoding='utf-8')
st902.head()
```

```python
# 定义目标股票
stock = "科大讯飞"
# 提取科大讯飞个股的信息，并合并为单个DataFrame
frame002230 = pd.concat([
    st902[st902['STOCK_NAME'] == stock], st903[st903['STOCK_NAME'] == stock],
    st904[st904['STOCK_NAME'] == stock], st905[st905['STOCK_NAME'] == stock],
    st906[st906['STOCK_NAME'] == stock]
],
                        axis=0)
frame002230
```

```python
# 设置day字段为索引
frame002230.set_index(r'DAY', inplace=True)
# 把原始数据中的开盘价、收盘价、最大最小、流通市值等指标转化为float类型，供接下来使用
frame002230[[
    "CURRENT_PRICE", "VOLUME", "TURNOVER", "MAXIMUM", "MINIMUM", "OPEN_TODAY",
    "CLOSE_YESTERDAY", "CURRENT_MARKET"
]] = frame002230[[
    "CURRENT_PRICE", "VOLUME", "TURNOVER", "MAXIMUM", "MINIMUM", "OPEN_TODAY",
    "CLOSE_YESTERDAY", "CURRENT_MARKET"
]].astype(float)
# 计算周期内的最高价
maximum_w = frame002230['MAXIMUM'].max()
print(maximum_w)
# 计算周期内的最低价
minimum_w = frame002230['MINIMUM'].min()
print(minimum_w)
# 计算最大振幅
amplitude = (maximum_w -
             minimum_w) / (frame002230.loc['20190902']['CLOSE_YESTERDAY'])
print('{0:.2f}%'.format(amplitude * 100))
```

```python
# 换算单位为，万手
volume_w = frame002230['VOLUME'].sum() / 10**4
print('%.2f万手' % volume_w)
# 换算单位为，亿元
turnover_w = frame002230['TURNOVER'].sum() / 10**8
print('本周交易额为%.2f亿元' % turnover_w)
```

```python
# 计算周期内的开盘价、收盘价，涨跌幅
open_w = frame002230.loc['20190902']['OPEN_TODAY']
close_w = frame002230.loc['20190906']['CURRENT_PRICE']
changemount_w = close_w - open_w
changepercent_w = changemount_w / (
    frame002230.loc['20190902']['CLOSE_YESTERDAY'])
print('本周开盘{0}元，收盘{1}元，涨跌{2:.2f}元，涨跌幅{3:.2f}%'.format(open_w, close_w,
                                                      changemount_w,
                                                      changepercent_w * 100))
```

```python
# 选取最后一天的流通市值为最新指标
market_w = frame002230.loc['20190906']['CURRENT_MARKET']
print(f'最新的流通市值为{market_w/10**8:.2f}亿元')
```

```python
# 这里我们以分析人工智能概念板块的股票为例，进行展示如何进行股票的批量分析以及报表生成。
AI = concept[concept['CONCEPT_NAME'] == '人工智能']
AI.shape
AI.head()
```

```python
# step 1: 把全部股票交易数据进行合并
frame_stock = pd.concat([st902, st903, st904, st905, st906], axis=0)
# step 2: 根据人工智能板块的股票池，筛选出归属于人工智能板块的股票交易信息
frame_AI = pd.merge(frame_stock,
                    AI,
                    on=['STOCK_NAME', 'STOCK_CODE'],
                    how='inner')
# step 3: 初步清洗：去除‘-’，并进行排序，按照时间和stock_code进行
frame_AI.replace('-', np.nan, inplace=True)
frame_AI.sort_values(['STOCK_CODE', 'DAY'], inplace=True)
# step 4： 更改关键指标为float类型，供后续计算
frame_AI[[
    "CURRENT_PRICE", "VOLUME", "TURNOVER", "MAXIMUM", "MINIMUM", "OPEN_TODAY",
    "CLOSE_YESTERDAY", "CURRENT_MARKET"
]] = frame_AI[[
    "CURRENT_PRICE", "VOLUME", "TURNOVER", "MAXIMUM", "MINIMUM", "OPEN_TODAY",
    "CLOSE_YESTERDAY", "CURRENT_MARKET"
]].astype(np.float)
pd.set_option('display.max_rows', 20)
frame_AI
```

```python
# 根据排序的规则，选择第一行指标（比如可以查看周期内的开盘价）
def top_most(arr):
    if arr.size > 0:
        return arr.iloc[0]
    else:
        return np.nan


# 根据排序的规则，选择最后一行指标（比如可以查看周期内的收盘价、最新市值等）
def last_most(arr):
    if arr.size > 0:
        return arr.iloc[-1]
    else:
        return np.nan


frame_AI_report = frame_AI.groupby(['STOCK_CODE', 'STOCK_NAME']).agg({
    'MAXIMUM': np.max,  # 计算最高价
    'MINIMUM': np.min,  # 最低价
    'VOLUME': np.sum,  # 累计成交量
    'TURNOVER': np.sum,  # 累计成交额
    'OPEN_TODAY': top_most,  # 周内开盘价
    'CURRENT_PRICE': last_most,  # 周内收盘价
    'CLOSE_YESTERDAY': top_most,  # 上周的收盘价
    'CURRENT_MARKET': last_most  # 周内最新流通市值
}).reset_index()
frame_AI_report
```

```python
# 重命名列
frame_AI_report.rename(columns={
    'MAXIMUM': 'maximum_w',
    'MINIMUM': 'minimum_w',
    'VOLUME': 'volume_w',
    'TURNOVER': 'turnover_w',
    'OPEN_TODAY': 'open_w',
    'CURRENT_PRICE': 'close_w',
    'CLOSE_YESTERDAY': 'close_last_w',
    'CURRENT_MARKET': 'market_w'
},
                       inplace=True)
frame_AI_report
```

```python
# 计算最大振幅
frame_AI_report['amplitude_w'] = frame_AI_report.apply(
    lambda x: (x.maximum_w - x.minimum_w) / x.close_last_w
    if pd.notnull(x.maximum_w) and pd.notnull(x.minimum_w) and pd.notnull(
        x.close_last_w) else np.nan,
    axis=1)
# 计算周期内涨跌额
frame_AI_report['changemount_w'] = frame_AI_report.apply(
    lambda x: x.close_w - x.open_w
    if pd.notnull(x.close_w) and pd.notnull(x.open_w) else np.nan,
    axis=1)
# 计算周期内涨跌幅
frame_AI_report['changepercent_w'] = frame_AI_report.apply(
    lambda x: x.changemount_w / x.close_last_w
    if pd.notnull(x.close_last_w) and pd.notnull(x.changemount_w) else np.nan,
    axis=1)
frame_AI_report
```

```python
# 成交量单位：万手
frame_AI_report['volume_w'] = frame_AI_report['volume_w'] / 10**4
# 成交额单位：亿元
frame_AI_report['turnover_w'] = frame_AI_report['turnover_w'] / 10**8
# 流通市值单位：亿元
frame_AI_report['market_w'] = frame_AI_report['market_w'] / 10**8
frame_AI_report
```

```python
# 对于小数位，保留2位小数
frame_AI_report[[
    'maximum_w', 'minimum_w', 'volume_w', 'turnover_w', 'open_w', 'close_w',
    'close_last_w', 'market_w', 'changemount_w'
]] = frame_AI_report[[
    'maximum_w', 'minimum_w', 'volume_w', 'turnover_w', 'open_w', 'close_w',
    'close_last_w', 'market_w', 'changemount_w'
]].round(2)
# 对于部分数据，例如涨幅，转化为百分位数，并保留2位小数
frame_AI_report[[
    'changepercent_w', 'amplitude_w'
]] = frame_AI_report[['changepercent_w', 'amplitude_w']].round(4) * 100
pd.set_option('display.max_rows', 20)
frame_AI_report
```

```python
# 查看一周上涨在20%以上的股票
frame_AI_report[frame_AI_report['changepercent_w'] > 20]
```

```python
# 查看一周内呈下跌趋势的股票
frame_AI_report[frame_AI_report['changepercent_w'] < 0]
```
