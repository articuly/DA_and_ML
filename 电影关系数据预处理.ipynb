```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel('data_set/豆瓣电影数据.xlsx', sheet_name=0, header=0, index_col=0)
```

```python
df.head()
```

```python
# 数据清洗

data=df[['name', '导演', '主演']]
data.head()
```

```python
data.dropna(inplace=True)
```

```python
# 主演分列
data_yy=data['主演'].str.split('/', expand=True)
col_len1=len(data_yy.columns)
data_yy.columns=['yy'+str(i) for i in range(col_len1)]
data_yy.head()
```

```python
# 导演分裂
data_dy=data['导演'].str.split('/', expand=True)
col_len2=len(data_dy.columns)
data_dy.columns=['dy'+str(i) for i in range(col_len2)]
data_dy.head()
```

```python
# 数据合并
data2=data_dy.join(data_yy).join(data['name'])  # 横向合并
data2.head()
```

```python
data2[data2['yy64'].notnull()]  # 可以看看分出64个演员的数据 → 海角七号
```

```python
# 数据拆分，再合并
# 创建空DF
data_re=pd.DataFrame(columns=['电影名','导演','演员'])

# 提取列名
col_yy=data_yy.columns
col_dy=data_dy.columns
```

```python
for dy in col_dy:
    for yy in col_yy:
        data_i=data2[['name', dy, yy]].dropna()
        data_i.columns=['电影名','导演','演员']
        data_re=pd.concat([data_re, data_i])

data_re.head()
# 遍历数据后，得到一个导演与演员的关系数据，并做去重处理
# 这里index是有重复的，但作为过程数据可忽略
```

```python
data_re.shape
```

```python
# 汇总统计导演和演员的合作次数
result=data_re.groupby(['导演','演员']).count()
result.reset_index(inplace=True)
result.head()
```

```python
writer=pd.ExcelWriter(r'data_set/电影关系数据.xlsx')
result.to_excel(writer, 'sheet1')
writer.save()
print('finished!')
```
